먼저 Langchain의 OpenAI 모듈을 설치한다.

```python
LLM_PROVIDER = "openai" # claude 또는 gemini
TEMPERATURE=0.2
```

ChatOpenAI를 임포트 한다.

OPENAI 키는 비밀리에 다뤄야 하므로, 외부에서 입력받도록 한다.<BR> 이를 위해 외부입력을 받는 모듈을 임포트 한다.

```python
import getpass
```

키를 외부로부터 받아, OPENAI_API_KEY 변수에 둔다.

```python
API_KEY = getpass.getpass("YOUR KEY: ")
```

실험용이므로, 성능에 상관없이 가장 저렴한 모델을 호출


```python
if LLM_PROVIDER == "openai":
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(
        model="gpt-5-nano",
        temperature=TEMPERATURE,
        api_key=API_KEY,
    )


elif LLM_PROVIDER == "claude":
    from langchain_anthropic import ChatAnthropic

    llm = ChatAnthropic(
        model="claude-3-5-sonnet-latest",
        temperature=TEMPERATURE,
        api_key=API_KEY,
    )


elif LLM_PROVIDER == "gemini":
    from langchain_google_genai import ChatGoogleGenerativeAI

    llm = ChatGoogleGenerativeAI(
        model="gemini-2.0-flash",
        temperature=TEMPERATURE,
        api_key=API_KEY,
    )
```

프롬프트를 작성해 본다.

```python
prompt_input = "세계 최고의 대학 Top 10 알려줘"
```

엔진을 호출하고, 결과를 받는다.

```python
response = llm.invoke(prompt_input)
```

결과를 출력해 본다.

```python
print(response.content)
```

이번에는 명령 템플릿을 사용하여 호출해 본다.<BR> 먼저 프롬프트 템플릿을 임포트 한다.

```python
from langchain_core.prompts import PromptTemplate
```

템플릿을 작성한다.

```python
prompt_template = PromptTemplate.from_template("""너는 기자야. 이제 다음 질문에
{num_words} 단어로 대답해. 그리고 어투는 {tone}에 맞춰: {text}""")
```

질문 프롬프트를 작성한다.

```python
prompt_text = "한국에서 가장 아름다운 도시 10개 알려줘"
```

템플릿을 구성한다.

```python
prompt_input = prompt_template.format(
    text=prompt_text,
    num_words=100,
    tone="간결하게")
```

엔진을 호출하고, 결과를 받는다.

```python
response = llm.invoke(prompt_input)
```

결과를 출력해 본다.

```python
print(response.content )
```

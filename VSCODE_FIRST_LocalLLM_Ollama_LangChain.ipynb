```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="gemma3:4b",
    temperature=0
)

messages = [
    (
        "system",
        "너는 한국어로 짧고 쉽게 답하는 AI 어시스턴트다."
    )
]

print("Ollama + LangChain 로컬 챗봇 시작")
print("종료하려면 exit 입력")

while True:
    user_input = input("\n사용자: ")

    if user_input.lower() == "exit":
        print("종료합니다.")
        break

    messages.append(
        (
            "human",
            user_input
        )
    )

    response = llm.invoke(messages)

    print("\nAI:", response.content)

    messages.append(
        (
            "assistant",
            response.content
        )
    )
```

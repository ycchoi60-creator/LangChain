ollama 모델을 설정한다.

```python
MODEL = "gemma3:4b"
```

로컬 Ollama의 기본 URL을 설정한다.

```python
URL = "http://localhost:11434/api/chat"
```

명령어의 기본 골격 메시지를 구성한다.

```python
messages = [
    {
        "role": "system",
        "content": "너는 한국어로 짧고 쉽게 답하는 AI 어시스턴트다."
    }
]
```

루프를 돌며 챗봇을 운영한다.

```python
print("Ollama 로컬 챗봇 시작")
print("종료하려면 exit 입력")

while True:
    user_input = input("\n사용자: ")

    if user_input.lower() == "exit":
        print("종료합니다.")
        break

    messages.append(
        {
            "role": "user",
            "content": user_input
        }
    )

    response = requests.post(
        URL,
        json={
            "model": MODEL,
            "messages": messages,
            "stream": False
        }
    )

    answer = response.json()["message"]["content"]

    print("\nAI:", answer)

    messages.append(
        {
            "role": "assistant",
            "content": answer
        }
    )
```
